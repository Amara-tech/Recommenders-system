```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv("Dataset/tmdb_5000_credits.csv")
```

```python
movies_df = pd.read_csv("Dataset/tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies_df.head()
```

```python
credits_renamed = credits.rename(index=str, columns={"movie_id": "id"})
movies_df_merge = movies_df.merge(credits_renamed, on="id")
```

```python
movies_df_merge.head(5)
```

```python
movies = movies_df_merge.drop(columns=["homepage", "title_x", 'title_y', 'status', 'production_companies'])
movies.head()
```

```python
movies.info()
```

W = (Rv + Cm)/v+m.   
W = weighted Rating
R= average for the movie as a number from 0 to 10(mean) = (Rating)
v= number of votes for thr movie = (votes)
m=minimum votes required to be listed in the top 250(currently 3000)
C= the mean vote across the whole report (currently 6.9)

```python
v=movies['vote_count']
R=movies['vote_average']
C=movies['vote_average'].mean()
m=movies['vote_count'].quantile(0.7)
```

```python
movies['weighted_average']= (R*v) + (C*m) / (v + m)
```

```python
movies_sorted = movies.sort_values('weighted_average', ascending=True)
movies_sorted.head()
```

```python
movies_sorted[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(20)
```
