```python
ratings = pd.read_csv('../Dataset/Ratings.csv', encoding='latin-1', low_memory=False)
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id', 'Book-Rating':'rating'}, inplace=True)
```

```python
ratings['user_id'].value_counts()
```

```python
x = ratings['user_id'].value_counts() > 200
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings_with_book = ratings.merge(books, on='ISBN')
```

```python
ratings_with_book.head()
```

```python
number_rating = ratings_with_book.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
final_rating = ratings_with_book.merge(number_rating, on='title')
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(['user_id', 'title'], inplace=True)
```

```python
final_rating.isnull().sum()
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='ratings')
```
